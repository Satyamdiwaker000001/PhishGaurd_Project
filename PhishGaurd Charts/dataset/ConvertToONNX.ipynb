```python
# ==========================================
# ⚡ CONVERT TO TFLITE (Google's Native Format)
# ==========================================
import tensorflow as tf
import os
import shutil

# 1. SETUP PATHS
INPUT_MODEL = '/content/drive/MyDrive/Colab Notebooks/Dataset/phishguard_model_cnn.h5'
OUTPUT_TFLITE = '/content/drive/MyDrive/Colab Notebooks/Dataset/phishguard_brain.tflite'

print("🚀 Starting TFLite Conversion...")

try:
    # 2. LOAD MODEL
    print("⏳ Loading Keras Model...")
    model = tf.keras.models.load_model(INPUT_MODEL)

    # 3. CONVERT (Native Way)
    print("🔄 Converting to TFLite...")
    converter = tf.lite.TFLiteConverter.from_keras_model(model)

    # Optimization (Optional: Size aur chhota karne ke liye)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]

    tflite_model = converter.convert()

    # 4. SAVE FILE
    with open("temp_brain.tflite", 'wb') as f:
        f.write(tflite_model)

    # Drive mein move karo
    shutil.move("temp_brain.tflite", OUTPUT_TFLITE)

    print(f"\n🎉 SUCCESS! TFLite File Ban Gayi: {OUTPUT_TFLITE}")
    print("👉 Drive se 'phishguard_brain.tflite' download kar lo.")

except Exception as e:
    print(f"\n❌ Error: {e}")
```
